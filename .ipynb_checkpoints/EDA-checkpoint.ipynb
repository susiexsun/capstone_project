```python
import pymongo
from pymongo import MongoClient
import numpy as np
import pandas as pd
import json
```

```python
client = MongoClient()
twitter = client['twitter']
tweets = twitter['english']
docs = tweets.find({}, {'user.screen_name':1, 'text':1, 'user.followers_count': 1, 'user.friends_count': 1})
```

```python
doc_dict = {}

for doc in docs:
    user = doc.get('user').get('screen_name').encode('utf8', 'ignore')
    if user not in doc_dict: 
        followers = doc.get('user').get('followers_count')
        following = doc.get('user').get('friends_count')
        doc_dict[user] = followers, following
```

```python
df = pd.DataFrame(doc_dict).transpose()
```

```python
df.head()
```

```python
df = df.rename(columns={0: "followers", 1: "following"})
```

```python
df['followers_ratio'] = df['followers'] / df['following']
```

```python
df.head()
```

```python
df.sort(['followers_ratio'])
```

```python
df[(df['followers_ratio'] > 1.5) & (df['following'] > 100)]
```
